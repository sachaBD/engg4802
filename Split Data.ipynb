```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('data/raw_data.csv')
```

```python
# Split into 15 sections
num_splits = 15

split_indexes = np.linspace(0, len(data) - len(data) % num_splits, num_splits + 1, dtype=np.int)

split_data = []

for i in range(0, 15):
    split_data += [data[split_indexes[i] : split_indexes[i+1]]]
```

```python
# Take the moving average
mva_split_data = []
periods = [1, 10, 30, 45]

for period in periods:
    for d in split_data:
        mva_split_data += [ pd.DataFrame(d).rolling(window=period).mean() ]
```

```python
# Convert to dataframe
columns = []

for period in [1, 10, 30, 45]:
    for i, d in enumerate(split_data):
        columns += ['window_' + str(period) + '_' + str(i)]

data_table = pd.DataFrame(data=np.hstack(mva_split_data), columns=columns)

# Remove the MVA nans
data_table = data_table.iloc[max(periods):]
```

```python
# Check for nan
pd.isna(data_table).any().any()
```
